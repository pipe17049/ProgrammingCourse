# Execise 1 

Input: list of strings 
Output: List of each strings sizes

Example:

Input
["Hello", "world",":D"]
Ouput 
[5, 5 , 2]


```python
input_list = ["casa","sol","chatarra"]
ans = []
ans_2 = []
for word in input_list:
    ans.append(len(word))
print(ans)

for index in range(0, len(input_list)):
    ans_2.append(len(input_list[index]))
print(ans_2)
    
```
